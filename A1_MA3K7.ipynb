```python
import numpy as np
```

```python
## the first 100 Fibonacci Numbers modulo 10
F = np.zeros(100,dtype=int)
F[0] = 1
F[1] = 1
for i in range(2,100):
    F[i] = ( F[i-1] + F[i-2] ) % 10
    
## Finding the n specified by the Rubric
## i.e. finding n such that F_{n-1} % 10 = 1, F_{n} % 10 = 0 and F_{n+1} is odd
for i in range(2,100):
    if F[i-2] == 1 and F[i-1] == 0 and F[i]%2 == 1 :
        print(i)
        break 
```

```python
## Function to find the size of the loop of a number bracelet that starts with the numbers a,b
## Since there are 100 possible combinations of inputting digits 0 to 9 as a,b, the maximum test is 100 
def size_of_nb(a,b):
    seq = np.zeros(100,dtype=int)
    seq[0]=a
    seq[1]=b
    for j in range(2,100):
        seq[j] = ( seq[j-2] + seq[j-1] ) % 10
        if seq[j-1] == a and seq[j] == b:
            return j-1
            break

## Generate a 10x10 matrix where the (i,j) value of the matrix corresponds to the Number Bracelet generated by (i,j)
FNB = np.zeros((10,10),dtype=int)
for i in range(10):
    for j in range(10):
        FNB[i,j] = size_of_nb(i,j)
        
## Find all groups specified by the Rubric
## i.e. exclude even pairs (as such I chose to only check for when the first number in the pair is odd)
## i.e. exclude the sequence generated by (5,0),(5,5) and (0,5) (which is already excluded since 0 is even).
for i in range(5):
    for j in range(10):
        k = 2*i+1
        if FNB[k,j] <= 12:
            if k != 5 and j != 0:
                if k != 5 and j != 5:
                    print(k,j,"gives FNB of size",FNB[k,j])
```

```python
## Checking results of the Rubric
## i.e. checking that there are 6 distinct FNB sizes that add up to 100
FNB_sizes = np.unique(FNB)
print("The following are all the unique sizes of the FNB's:",FNB_sizes)
sum_of_sizes = sum(FNB_sizes)
print("These sizes sum up to",sum_of_sizes)
```
